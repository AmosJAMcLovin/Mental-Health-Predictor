```python
import os
from datetime import datetime
import numpy as np 
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
survey = pd.read_csv('Data/survey.csv')
```

```python
survey.head()
```

# Data Cleaning

```python
# removing columns not useful for our analysis
survey.drop(['Timestamp'], axis= 1,inplace = True)
```

```python
# the NaN values of the "self_employed" column will be filled with the largely dominant value: "No"
survey['self_employed'].fillna('No',inplace = True)
survey['self_employed'].value_counts()
```

```python
# Aggrigate value by gender
survey['Gender'].value_counts()
```

```python
# changing "Yes" and "No" values by 1 and 0 (to make plots possible)
survey.replace(('Yes', 'No'), (1, 0), inplace=True)

# Gender versus treatment
gender_treatment= survey.pivot_table(index="Gender",values="treatment")
gender_treatment.plot.bar(figsize=(10,8),fontsize = 14)

plt.xlabel("Gender",fontsize = 14)
plt.ylabel("Proportion of people who sought treatment",fontsize = 14)
plt.title("Gender Versus Treatment",fontsize = 20)

plt.savefig("Image/gender_vs_treatment", bbox_inches = "tight")
plt.tight_layout()
plt.show()
```

```python
# Replace NA values in work interfernce fild with"Don\'t know" 
survey['work_interfere'].fillna('Don\'t know', inplace = True)
survey['work_interfere'].value_counts()
```

```python
# Categorizing age to make ready for analysis
cut_points = [17,30,50,75]
label_names = ["Young Adult","Adult","Senior"]
survey["Age_categories"] = pd.cut(survey["Age"],cut_points,labels=label_names)
```

```python
# plot treatmnet by age group
treatment_age = survey.pivot_table(index="Age_categories",values='treatment')
treatment_age.plot.bar(figsize=(10,8),fontsize = 14)

plt.xlabel("Age Group",fontsize = 14)
plt.ylabel("Proportion of people who sought treatment",fontsize = 14)
plt.title("Age Versus Treatment",fontsize = 20)

plt.savefig("Image/age_vs_treatment", bbox_inches = "tight")
plt.tight_layout()
plt.show()
```

```python
survey["Age"].describe()
```

```python
# Some values in the "Age" column are not realistic or completely no sense (like negative values or positive values above 150).
#Values not between 18 and 73 will be filled by the median value of the column.
survey['Age'][survey['Age']<18] = survey['Age'].median()
survey['Age'][survey['Age']>73] = survey['Age'].median()






```

## Normalizing the data for the ML Analysis 









```python
def create_dummies(df,column_name):
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df
survey = create_dummies(survey,"Age_categories")
survey = create_dummies(survey,"Gender")

# creating dummy columns for other columns
cols = ['family_history','work_interfere', 'care_options','benefits', 'wellness_program','anonymity',\
        'leave','mental_health_consequence', 'phys_health_consequence']
for column in cols:
    survey = create_dummies(survey,column)
    
display(survey.head())
display(survey.shape)
survey.to_csv(path_or_buf="Normalized_df.csv")
```

```python
# Try to see reletionships by ploting corroletion matrix
corrrelation=survey.corr()
corrrelation
```

```python
# plot the correlation matrix
corrMatrix=corrrelation.corr()
sns.heatmap(corrMatrix,annot=True)
plt.savefig("corrmatrix.png", bbox_inches="tight")
plt.show()
```
